```python
import matplotlib.pyplot as plt
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
```

```python
plt.pie([1])
plt.show()
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
plt.pie(class1_stu,labels=classes)
plt.show()
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
plt.pie(class1_stu,labels=classes,explode=[0,.1,0,0],colors=["g","r","y","b"],autopct="%0.2f%%")
plt.show()
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
plt.pie(class1_stu,labels=classes,explode=[0,.1,0,0],colors=["g","r","y","b"],autopct="%0.2f%%",shadow=True,radius=1.4)
plt.show()
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
plt.pie(class1_stu,labels=classes,explode=[0,.1,0,0],colors=["g","r","y","b"],autopct="%0.2f%%",shadow=True,radius=1.4,startangle=270)
plt.show()
```

```python
classes=["python","r","java","machine learning"]
class1_stu=[45,14,35,25]
plt.pie(class1_stu,labels=classes,explode=[0,.1,0,0],colors=["g","r","y","b"],autopct="%0.2f%%",shadow=True,radius=1.4,startangle=270,textprops={"fontsize":15})
plt.show()
```
