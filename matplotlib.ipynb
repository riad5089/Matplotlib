```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style  
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width=.3,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--")
```

```python
plt.bar(classes,class1_students,width=.3,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=.3,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")

#plt.grid(color='k')
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")
plt.title("Bar chart of iaip class",fontsize=18)
plt.xlabel("class",fontsize=5)
plt.ylabel("no of student",fontsize=15)
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,color="g",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")
plt.bar(classes,class2_students,color="r",edgecolor="g",linewidth=5,alpha=0.9,linestyle="--",label="class1_stu")
plt.title("Bar chart of iaip class",fontsize=18)
plt.xlabel("class",fontsize=5)
plt.ylabel("no of student",fontsize=15)
```
